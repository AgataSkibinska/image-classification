### Images analysis

```python
np.unique(X_train)
```

Note: Matrices' values are already normalized into binary values.

```python
def show_examples(X, y, label):
    n_examples = 3
    idxs_filtered = np.where(y == label)[0]
    
    X_filtered = X[idxs_filtered][:n_examples]
    
    if len(X_filtered) < n_examples:
        n_examples = len(X_filtered)
    X_reshaped = np.reshape(X_filtered, (-1, 56, 56))
    
    
    fig, axes = plt.subplots(1, n_examples, constrained_layout=True)
    # fig.subplots_adjust(top=1.5)
    fig.suptitle(f'Examples from class {label}')
    for i in range(n_examples):
        if n_examples == 1:
            axes.imshow(X_reshaped[i])
        else:
            axes[i].imshow(X_reshaped[i])
    fig.savefig(f'images/examples_label_{label}.png',  bbox_inches='tight')
    plt.show()
```

```python
for label in range(num_classes):
    show_examples(X_train, y_train, label)
```

```python
for label in range(num_classes):
    show_examples(X_train, y_train, label)
```
